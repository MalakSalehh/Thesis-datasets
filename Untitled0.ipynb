**5. Prompt templates**

```python
def build_baseline_prompt(essay_text):
    return f"""
You are an expert essay grader evaluating student essays.

Essay Prompt:
{essay_prompt_text}

{rubric_text}

Calibration Examples:
{calibration_text}

Essay to Evaluate:
{essay_text}

Instructions:
1. Predict the final score (2–12).
2. Predict the category (Low, Medium, High).
3. Explain your reasoning briefly.

Output Format:
Final Score:
Category:
Reasoning:
""".strip()
```

```python
def get_response_with_history(messages, model_name="openai/gpt-oss-120b", temperature=0.3, max_tokens=1024):
    response = client.chat.completions.create(
        model=model_name,
        messages=messages,
        temperature=temperature,
        max_tokens=max_tokens
    )
    return response.choices[0].message.content
```

```python
def build_followup_prompt(essay_text):
    return f"""
Now evaluate the following essay using the same rubric and calibration context.

Essay to Evaluate:
{essay_text}

Instructions:
1. Predict the final score (2–12).
2. Predict the category (Low, Medium, High).
3. Explain your reasoning briefly.

Output Format:
Final Score:
Category:
Reasoning:
""".strip()
```

```python
def build_mts_prompt(essay_text):
    return f"""
You are an expert essay grader evaluating student essays.

Essay Prompt:
{essay_prompt_text}

{rubric_text}

Calibration Examples:
{calibration_text}

Essay to Evaluate:
{essay_text}

Instructions:
Step 1: Content Quality
- Does the essay present a clear opinion?
- Does it include relevant ideas and support?
Score from 1–6.

Step 2: Organization
- Is the essay logically structured?
- Are ideas connected clearly?
Score from 1–6.

Step 3: Language & Clarity
- Is the writing understandable?
- Are grammar or spelling issues affecting meaning?
Score from 1–6.

Step 4: Final Score
- Calculate the average of the three trait scores.
- Multiply by 2.
- Round to the nearest integer.

Step 5: Category
- Low = 2–5
- Medium = 6–8
- High = 9–12

Output Format:
Content Score:
Organization Score:
Language Score:
Average:
Final Score:
Category:
Reasoning:
""".strip()
```

**6. API call function**

```python
def get_response(prompt, model_name="openai/gpt-oss-120b", temperature=0.3, max_tokens=1024):
    response = client.chat.completions.create(
        model=model_name,
        messages=[
            {"role": "system", "content": "You are an expert essay grader. Follow the rubric strictly."},
            {"role": "user", "content": prompt}
        ],
        temperature=temperature,
        max_tokens=max_tokens
    )
    return response.choices[0].message.content
```

**7. Parsing function**

```python
import re
import pandas as pd

def clean_output_text(text):
    if pd.isna(text):
        return None
    text = str(text)
    text = text.replace("**", "")
    text = text.replace("\\n", "\n")
    return text.strip()


def extract_final_score(text):
    if pd.isna(text):
        return None

    text = clean_output_text(text)

    # Get only the line that contains "Final Score:"
    match = re.search(r"Final Score:\s*(.*)", text, re.IGNORECASE)
    if not match:
        return None

    score_line = match.group(1).strip()

    # Find all numbers on that line
    nums = re.findall(r"\d+(?:\.\d+)?", score_line)
    if not nums:
        return None

    # Prefer the last integer-looking number between 2 and 12
    for value in reversed(nums):
        num = float(value)
        if num.is_integer():
            score = int(num)
            if 2 <= score <= 12:
                return score

    # Fallback: round the last numeric value
    score = int(round(float(nums[-1])))
    return score if 2 <= score <= 12 else None


def extract_category(text):
    if pd.isna(text):
        return None

    text = clean_output_text(text)

    match = re.search(r"Category:\s*(Low|Medium|High)", text, re.IGNORECASE)
    return match.group(1).capitalize() if match else None
```

**8. Experiment runner**

```python
def run_experiment(df_subset, prompt_builder, config_name, pause_seconds=1):
    results = []

    for _, row in df_subset.iterrows():
        essay_id = row["essay_id"]
        essay_text = row["essay"]
        human_score = row["domain1_score"]
        human_category = row["score_category"]

        prompt = prompt_builder(essay_text)
        output = get_response(prompt)

        results.append({
            "configuration": config_name,
            "essay_id": essay_id,
            "human_score": human_score,
            "human_category": human_category,
            "model_output": output
        })

        print(f"{config_name} | Done Essay {essay_id}")
        time.sleep(pause_seconds)

    results_df = pd.DataFrame(results)
    results_df["predicted_score"] = results_df["model_output"].apply(extract_final_score)
    results_df["predicted_category"] = results_df["model_output"].apply(extract_category)

    return results_df
```

```python
results_config4["clean_output"] = results_config3["model_output"].apply(clean_output_text)
results_config4["predicted_score"] = results_config3["clean_output"].apply(extract_final_score)
results_config4["predicted_category"] = results_config3["clean_output"].apply(extract_category)

results_config4[["essay_id", "human_score", "predicted_score", "predicted_category", "clean_output"]]
```

```python
def compute_metrics(results_df):
    clean_df = results_df.dropna(subset=["predicted_score"]).copy()

    qwk = cohen_kappa_score(
        clean_df["human_score"],
        clean_df["predicted_score"],
        weights="quadratic"
    )

    pcc, _ = pearsonr(
        clean_df["human_score"],
        clean_df["predicted_score"]
    )

    mae = mean_absolute_error(
        clean_df["human_score"],
        clean_df["predicted_score"]
    )

    return {
        "Essays (n)": len(clean_df),
        "QWK": round(qwk, 4),
        "PCC": round(pcc, 4),
        "MAE": round(mae, 4)
    }
```
